```python
# Decisión sobre los valores NA
df_limpio = df_accidentes.dropna()

df_accidentes['CANT_HERIDOS_EN _SITIO_ACCIDENTE'].fillna(df_accidentes['CANT_HERIDOS_EN _SITIO_ACCIDENTE'].mean(), inplace=True)
df_accidentes['CANT_MUERTOS_EN _SITIO_ACCIDENTE'].fillna(df_accidentes['CANT_MUERTOS_EN _SITIO_ACCIDENTE'].mean(), inplace=True)

# Mostrar el DataFrame limpio
df_limpio.head()
```

### Interpretación del gráfico
En esta visualización, se puede observar cómo las variables interactúan y su comportamiento a lo largo del tiempo o bajo diferentes condiciones. Aquí podemos analizar las tendencias generales, los picos inusuales o las relaciones clave entre las variables representadas en la gráfica.

```python
acc_dia_gravedad = df_accidentes.groupby(['DIA_ACCIDENTE', 'GRAVEDAD_ACCIDENTE'])['CANTIDAD_ACCIDENTES'].sum().reset_index()

acc_dia_total = acc_dia_gravedad.groupby('DIA_ACCIDENTE')['CANTIDAD_ACCIDENTES'].sum().reset_index()
acc_dia_total = acc_dia_total.sort_values(by='CANTIDAD_ACCIDENTES', ascending=False)
orden = acc_dia_total['DIA_ACCIDENTE'].tolist()

plt.figure(figsize=(12, 6))
sns.barplot(data=acc_dia_gravedad, x='DIA_ACCIDENTE', y='CANTIDAD_ACCIDENTES', hue='GRAVEDAD_ACCIDENTE', order =orden , palette=["#281aa3", "#20138e", "#4234bb"])

plt.ylim(0, 2800)
plt.xlabel('Día de Accidente')
plt.ylabel('Total de Accidentes')
plt.title('Total de Accidentes por Día y Gravedad')
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend(loc='upper right')

plt.show()
```
